## 1. Importación de librerías

## 2. Descarga automática de datos desde el Banco Mundial

```python
paises_latam_iso = ['CO','BR','MX','AR','CL','PE','EC','BO','VE','PY','UY','CR','GT','HN','SV']
paises_str = ';'.join(paises_latam_iso)

def descargar_indicador(codigo, nombre_col):
    url = (f'https://api.worldbank.org/v2/country/{paises_str}/indicator/{codigo}'
           f'?format=json&per_page=2000&mrv=30')
    resp = requests.get(url, timeout=30)
    datos = resp.json()
    registros = []
    for item in datos[1]:
        registros.append({
            'Country Name': item['country']['value'],
            'Country Code': item['countryiso3code'],
            'Year': int(item['date']),
            nombre_col: item['value']
        })
    df = pd.DataFrame(registros)
    print(f'📂 {nombre_col}: {df.shape[0]} filas x {df.shape[1]} columnas')
    return df

df_gini    = descargar_indicador('SI.POV.GINI', 'Gini')
df_pobreza = descargar_indicador('SI.POV.DDAY', 'Pobreza_Extrema_Pct')
print('\n✅ Ambos datasets cargados correctamente')
```

## 4. Estadísticas descriptivas — Dataset 1: Coeficiente de Gini

Media, mediana, moda, desviación estándar, nulos, máximo y mínimo de **todas las columnas**.

```python
print('=' * 60)
print('ESTADÍSTICAS DESCRIPTIVAS — DATASET GINI')
print('=' * 60)

resumen_gini = []
for col in df_gini.columns:
    serie = df_gini[col]
    fila = {'Columna': col, 'Tipo': str(serie.dtype), 'Nulos': int(serie.isnull().sum())}
    if pd.api.types.is_numeric_dtype(serie):
        fila['Media']     = round(serie.mean(), 4)
        fila['Mediana']   = round(serie.median(), 4)
        fila['Moda']      = serie.mode()[0] if not serie.mode().empty else None
        fila['Desv. Std'] = round(serie.std(), 4)
        fila['Máximo']    = serie.max()
        fila['Mínimo']    = serie.min()
    else:
        fila['Media'] = fila['Mediana'] = fila['Desv. Std'] = 'N/A (texto)'
        fila['Moda']   = serie.mode()[0] if not serie.mode().empty else 'N/A'
        fila['Máximo'] = serie.max()
        fila['Mínimo'] = serie.min()
    resumen_gini.append(fila)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
display(pd.DataFrame(resumen_gini))
```

## 8. 🔎 Problemas de calidad de datos encontrados

### 8.1 Valores nulos (datos faltantes)
Este es el problema más significativo en ambos datasets. Los datos del Banco Mundial presentan una gran cantidad de valores nulos (`NaN`), especialmente en años anteriores a 1990 y en años recientes (2022–2023) donde el reporte aún no está completo. Países como Venezuela tienen registros muy incompletos por falta de reporte oficial.

### 8.2 Cobertura desigual entre países
No todos los países tienen el mismo número de registros válidos. Colombia y Brasil tienen datos más completos que Honduras o Paraguay, lo que puede introducir sesgo al comparar países directamente.

### 8.3 Alta proporción de nulos en los indicadores principales
Las columnas `Gini` y `Pobreza_Extrema_Pct` tienen muchos nulos porque estos indicadores no se miden todos los años; las encuestas de hogares se realizan cada 2 a 5 años según el país.

### 8.4 Frecuencia de medición irregular
Algunos países reportan datos cada año y otros cada 3 o 5 años, lo que hace que las series de tiempo sean irregulares y difíciles de comparar sin interpolación previa.

### 8.5 Tipo de dato de la columna Year
La columna `Year` viene como entero (`int64`), pero en análisis de series de tiempo conviene convertirla a tipo fecha (`datetime`) para facilitar graficaciones y operaciones temporales.

### 8.6 Ausencia de datos recientes para Venezuela
Venezuela no ha reportado datos oficiales al Banco Mundial desde 2006 para el índice Gini y desde 2019 para pobreza extrema, lo que la excluye prácticamente de cualquier comparación regional reciente.
